```python
# --- Step 1: Get median difference ---
med1 = np.median(data[data["Year"] <= 2015]["Notifications"])
med2 = np.median(data[data["Year"] > 2015]["Notifications"])
diff = med2 - med1
print(f"Median Difference: {diff:.2f}")

# --- Step 2: Bootstrap CI for median difference ---
n_boot = 1000
boot_diffs = []

group1 = data[data["Year"] <= 2015]["Notifications"].to_numpy()
group2 = data[data["Year"] > 2015]["Notifications"].to_numpy()

for _ in range(n_boot):
    boot1 = np.random.choice(group1, size=len(group1), replace=True)
    boot2 = np.random.choice(group2, size=len(group2), replace=True)
    boot_diffs.append(np.median(boot2) - np.median(boot1))

# Confidence interval
ci_low, ci_high = np.percentile(boot_diffs, [2.5, 97.5])
print(f"95% Confidence Interval: [{ci_low:.2f}, {ci_high:.2f}]")

# --- Step 3: Mann-Whitney U Test (non-parametric) ---
stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')
print(f"MannWhitney U Test: U={stat:.2f}, p-value={p_value:.4f}")
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Carregar dados

df_raca = df_raca[df_raca['CS_RACA'] != 'Ignorado']

def arima_projection(df_r, raca, forecast_year=2035):
    df_r = df_r.sort_values('NU_ANO')
    ts = df_r.set_index(pd.RangeIndex(len(df_r)))['ones']  # ndice sequencial
    
    # Ajustar ARIMA (p,d,q) = (1,1,1)
    model = ARIMA(ts, order=(1,1,1))
    result = model.fit()
    
    print(result.summary())
    
    # Previso at forecast_year
    n_forecast = forecast_year - df_r['NU_ANO'].max()
    forecast = result.get_forecast(steps=n_forecast)
    forecast_df = forecast.summary_frame()
    
    # Criar dataframe de projeo com anos corretos
    proj_years = list(range(df_r['NU_ANO'].max()+1, forecast_year+1))
    proj_df = pd.DataFrame({
        'NU_ANO': proj_years,
        'cases': forecast_df['mean'].values,
        'lower_ci': forecast_df['mean_ci_lower'].values,
        'upper_ci': forecast_df['mean_ci_upper'].values
    })
    
    # Histrico
    hist_df = df_r[['NU_ANO','ones']].rename(columns={'ones':'cases'})
    
    return hist_df, proj_df
```

```python
# Rodar para cada raa/cor e plotar
racas = df_raca['CS_RACA'].unique()

plt.figure(figsize=(12,8))

for raca in racas:
    print(raca)
    df_r = df_raca[df_raca['CS_RACA']==raca]
    hist_df, proj_df = arima_projection(df_r, raca)
    
    plt.plot(hist_df['NU_ANO'], hist_df['cases'], label=f'{raca} (observado)')
    plt.plot(proj_df['NU_ANO'], proj_df['cases'], linestyle='--', label=f'{raca} (projetado)')
    plt.fill_between(proj_df['NU_ANO'], proj_df['lower_ci'], proj_df['upper_ci'], alpha=0.2)
    
plt.xlabel('Ano')
plt.ylabel('Nmero de casos de TB')
plt.title('Projeo de casos de TB por raa/cor usando ARIMA')
plt.legend()
plt.show()
```

```python

# Remover ignorados
df_clean = df_raca[df_raca['CS_RACA'] != 'Ignorado']

# Funo para rodar ARIMA e projetar
def arima_forecast(df_r, order, forecast_year=2035):
    ts = df_r.set_index(pd.RangeIndex(len(df_r)))['ones']
    model = ARIMA(ts, order=order)
    result = model.fit()
    
    n_forecast = forecast_year - df_r['NU_ANO'].max()
    forecast = result.get_forecast(steps=n_forecast)
    forecast_df = forecast.summary_frame()
    
    proj_years = list(range(df_r['NU_ANO'].max()+1, forecast_year+1))
    proj_df = pd.DataFrame({
        'NU_ANO': proj_years,
        'cases': forecast_df['mean'].values,
        'lower_ci': forecast_df['mean_ci_lower'].values,
        'upper_ci': forecast_df['mean_ci_upper'].values
    })
    
    hist_df = df_r[['NU_ANO','ones']].rename(columns={'ones':'cases'})
    return hist_df, proj_df, result

# Dicionrio de ordens candidatas para cada raa/cor
orders_compare = {
    "Branca": [(1,1,1), (2,1,2)],
    "Preta": [(1,1,0), (2,1,1)],
    "Amarela": [(0,1,1), (1,1,1)],
    "Parda": [(1,1,1), (2,1,1)],
    "Indgena": [(1,1,1), (1,1,2)]
}

# Loop para cada grupo
for raca, orders in orders_compare.items():
    df_r = df_clean[df_clean['CS_RACA'] == raca].sort_values('NU_ANO')
    print(f"\n================= {raca} =================")
    
    hist, proj1, res1 = arima_forecast(df_r, orders[0], forecast_year=2035)
    _, proj2, res2 = arima_forecast(df_r, orders[1], forecast_year=2035)
    
    # Mostrar sumrios dos dois modelos
    print(f"\n=== Sumrio ARIMA{orders[0]} ===")
    print(res1.summary())
    print(f"\n=== Sumrio ARIMA{orders[1]} ===")
    print(res2.summary())
    
    # Plot
    plt.figure(figsize=(12,6))
    plt.plot(hist['NU_ANO'], hist['cases'], label='Observado', marker='o')
    
    plt.plot(proj1['NU_ANO'], proj1['cases'], linestyle='--', color='blue', label=f'ARIMA{orders[0]}')
    plt.fill_between(proj1['NU_ANO'], proj1['lower_ci'], proj1['upper_ci'], color='blue', alpha=0.2)
    
    plt.plot(proj2['NU_ANO'], proj2['cases'], linestyle='--', color='red', label=f'ARIMA{orders[1]}')
    plt.fill_between(proj2['NU_ANO'], proj2['lower_ci'], proj2['upper_ci'], color='red', alpha=0.2)
    
    plt.title(f"Projeo de casos de TB - Raa/Cor: {raca}")
    plt.xlabel("Ano")
    plt.ylabel("Nmero de casos de TB")
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings
warnings.filterwarnings("ignore")


# Funo com IC via bootstrap
def exp_smoothing_forecast(df_r, forecast_year=2035, n_sim=1000, alpha=0.05):
    df_r = df_r.sort_values('NU_ANO')
    ts = df_r.set_index('NU_ANO')['ones']
    
    # Ajustar modelo Holt-Winters (tendncia aditiva)
    model = ExponentialSmoothing(ts, trend='mul', seasonal=None)
    result = model.fit()
    
    # Previso pontual
    n_forecast = forecast_year - ts.index.max()
    forecast = result.forecast(n_forecast)
    proj_years = list(range(ts.index.max()+1, forecast_year+1))
    
    # Bootstrapping para IC
    resid = ts - result.fittedvalues
    sims = []
    for _ in range(n_sim):
        future = result.forecast(n_forecast)
        noise = np.random.choice(resid, size=n_forecast, replace=True)
        sims.append(future.values + noise)
    sims = np.array(sims)
    
    lower = np.percentile(sims, 100*alpha/2, axis=0)
    upper = np.percentile(sims, 100*(1-alpha/2), axis=0)
    
    proj_df = pd.DataFrame({
        'NU_ANO': proj_years,
        'cases': forecast.values,
        'lower_ci': lower,
        'upper_ci': upper
    })
    
    hist_df = ts.reset_index().rename(columns={'ones':'cases'})
    
    return hist_df, proj_df, result

# Rodar para todas as raas/cores


racas = df_clean['CS_RACA'].unique()

for raca in racas:
    df_r = df_clean[df_clean['CS_RACA'] == raca]
    hist, proj, model = exp_smoothing_forecast(df_r, forecast_year=2035)
    
    print(raca,round(proj.cases.mean()), round(proj.cases.sum()))
    
    # Plot
    plt.figure(figsize=(10,6))
    plt.plot(hist['NU_ANO'], hist['cases'], label='Observado', marker='o')
    plt.plot(proj['NU_ANO'], proj['cases'], linestyle='--', label='Projetado')
    plt.fill_between(proj['NU_ANO'], proj['lower_ci'], proj['upper_ci'], alpha=0.2, label='IC 95%')
    plt.title(f"Projeo de casos de TB - Raa/Cor: {raca} (Exponential Smoothing)")
    plt.xlabel("Ano")
    plt.ylabel("Nmero de casos de TB")
    plt.legend()
    plt.show()
```
